## Carregar os dados p df

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

# Pegar primeiro esse GaCo01_01
file_path = 'physionet.org/files/gaitpdb/1.0.0/GaCo01_01.txt'

# Arrumar/Renomear as colunas
# vai ser L de Left e R de Right
column_names = [
    'Time',
    'L_VGRF_1', 'L_VGRF_2', 'L_VGRF_3', 'L_VGRF_4', 'L_VGRF_5', 'L_VGRF_6',
    'L_VGRF_7', 'L_VGRF_8',
    'R_VGRF_1', 'R_VGRF_2', 'R_VGRF_3', 'R_VGRF_4', 'R_VGRF_5', 'R_VGRF_6',
    'R_VGRF_7', 'R_VGRF_8',
    'Total_L_Foot_Force',
    'Total_R_Foot_Force'
]

df = pd.read_csv(file_path, sep='\s+', header=None, names=column_names)

print("\nHead\n")
print(df.head())

print("\nInfo\n")
print(df.info())

print("\nShape\n")
print(df.shape)

print("\nColumns\n")
print(df.columns)
```

#### Dados Demográficos

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

# Pegar o demographics.txt
file_path = 'physionet.org/files/gaitpdb/1.0.0/demographics.txt'

df = pd.read_csv(file_path, sep='\s+')

print("\nHead\n")
print(df.head())

print("\nInfo\n")
print(df.info())

print("\nShape\n")
print(df.shape)

print("\nColumns\n")
print(df.columns)
```
