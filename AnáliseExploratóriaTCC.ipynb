### Análise geral de vgrf

```python
import os

data_dir = "physionet.org/files/gaitpdb/1.0.0/"

column_names = [
    'Time',
    'L_VGRF_1', 'L_VGRF_2', 'L_VGRF_3', 'L_VGRF_4', 'L_VGRF_5', 'L_VGRF_6', 'L_VGRF_7', 'L_VGRF_8',
    'R_VGRF_1', 'R_VGRF_2', 'R_VGRF_3', 'R_VGRF_4', 'R_VGRF_5', 'R_VGRF_6', 'R_VGRF_7', 'R_VGRF_8',
    'Total_L_Foot_Force',
    'Total_R_Foot_Force'
]

def analyze_and_plot_vgrf(file_path):

  df = pd.read_csv(file_path, sep='\s+', header=None, names=column_names)

  file_name = os.path.basename(file_path).replace('.txt', '')

  # Plotar a série temporal da força de reação vertical do solo (VGRF) para o pé esquerdo
  plt.figure(figsize=(12, 6))
  sns.lineplot(data=df, x='Time', y='Total_L_Foot_Force', color='blue')
  plt.title(f'Série Temporal da Força de Reação Vertical do Solo (Pé Esquerdo) - {file_name}' )
  plt.xlabel('Tempo (segundos)')
  plt.ylabel('Força (Newtons)')
  plt.grid(True)
  #plt.savefig(f'vgrf_left_foot_timeseries_{file_name}.png') #Salvar nn senão fica mt coisa salva aqui
  plt.show
  #plt.close()
  print(f"Gráfico 'vgrf_left_foot_timeseries_{file_name}.png' gerado com sucesso.")

  # Plotar a série temporal da força de reação vertical do solo (VGRF) para o pé direito
  plt.figure(figsize=(12, 6))
  sns.lineplot(data=df, x='Time', y='Total_R_Foot_Force', color='red')
  plt.title(f'Série Temporal da Força de Reação Vertical do Solo (Pé Direito) - {file_name}' )
  plt.xlabel('Tempo (segundos)')
  plt.ylabel('Força (Newtons)')
  plt.grid(True)
  #plt.savefig(f'vgrf_right_foot_timeseries_{file_name}.png')
  plt.show()
  #plt.close()
  print(f"Gráfico 'vgrf_right_foot_timeseries_{file_name}.png' gerado com sucesso.")

analyze_and_plot_vgrf(os.path.join(data_dir, "GaCo01_01.txt"))
```

## Análise de Velocidade de Marcha por Estágio Hoehn & Yahr (Pacientes com Parkinson)

[Sobre a escala Hoehn and Yahr](https://www.physio-pedia.com/Hoehn_and_Yahr_Scale)

### Introdução
Hoehn e Yahr foi publicado originalmente em 1967 no periódico  Neurology  por Margaret Hoehn e Melvin Yahr e incluía os estágios 1 a 5. Em 1983, Larsen et al publicaram um artigo intitulado "Questões teóricas e práticas na avaliação de déficits e terapia no parkinsonismo", no qual incluíram uma modificação da escala original de Hoehn e Yahr. Eles adicionaram algumas pequenas alterações que não alteraram substancialmente a classificação básica publicada em 1967. Esta escala é usada junto com a Escala Unificada de Avaliação da Doença de Parkinson para uma melhor avaliação da doença de Parkinson.

### Propósito
A escala tem sido utilizada para o estadiamento da incapacidade funcional associada à doença de Parkinson. Ela ajuda a descrever a progressão da doença em diferentes estágios, permitindo-nos mensurar a gravidade do caso.

image.png

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "physionet.org/files/gaitpdb/1.0.0/demographics.html"

df = pd.read_html(file_path, header=0)

df = dfs[0]

#pegar só a galera PD
df_pd = df[df["Group"] == 'PD'].copy()

df_pd.rename(columns={'Speed_01 (m/sec)': 'Speed_01'}, inplace=True)

#colocar p os valores serem numéricos
df_pd["HoehnYahr"] = pd.to_numeric(df_pd["HoehnYahr"], errors='coerce')
df_pd["Speed_01"] = pd.to_numeric(df_pd["Speed_01"], errors='coerce')

df_pd.dropna(subset=["HoehnYahr", "Speed_01"], inplace=True)

print("\n### Estatísticas de Velocidade de Marcha por Estágio Hoehn & Yahr (Pacientes com Parkinson):\n")
print(df_pd.groupby("HoehnYahr")["Speed_01"].describe().to_markdown(numalign="left", stralign="left"))

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_pd, x="HoehnYahr", y="Speed_01", palette="viridis")
sns.stripplot(data=df_pd, x="HoehnYahr", y="Speed_01", color="black", size=3, jitter=True)
plt.title("Velocidade de Marcha por Estágio Hoehn & Yahr (Pacientes com Parkinson)")
plt.xlabel("Estágio Hoehn & Yahr")
plt.ylabel("Velocidade de Marcha (m/s)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
#plt.savefig("hoehn_yahr_gait_speed_distribution.png")
plt.show()
#plt.close()
print("Gráfico \'hoehn_yahr_gait_speed_distribution.png\' gerado com sucesso.")
```

## Análise da Velocidade de Marcha por Grupo e Condição de Tarefa

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "physionet.org/files/gaitpdb/1.0.0/demographics.html"

dfs = pd.read_html(file_path, header=0)

df = dfs[0]

df.rename(columns={
    "Speed_01 (m/sec)": "Speed_Normal",
    "Speed_10": "Speed_DualTask" #Speed_10 pelo que entendi é tarefa dupla
}, inplace=True)

df["Speed_Normal"] = pd.to_numeric(df["Speed_Normal"], errors='coerce')
df["Speed_DualTask"] = pd.to_numeric(df["Speed_DualTask"], errors='coerce')

df.dropna(subset=["Speed_Normal", "Speed_DualTask"], inplace=True)

df_filtered = df[df["Group"].isin(['PD', 'CO'])].copy()

# Transformar os dados para o formato long
df_plot = df_filtered[["Group", "Speed_Normal", "Speed_DualTask"]].melt(id_vars="Group",
                                                                        var_name="Task_Condition",
                                                                        value_name="Gait_Speed")

df_plot["Task_Condition"] = df_plot["Task_Condition"].map({
        "Speed_Normal": "Caminhada Normal",
        "Speed_DualTask": "Tarefa Dupla"
})

print("\n### Estatísticas de Velocidade de Marcha por Grupo e Condição de Tarefa:\n")
print(df_plot.groupby(["Group", "Task_Condition"])["Gait_Speed"].describe().to_markdown(numalign="left", stralign="left"))

# Comparação da caminhada normal e tarefa dupla
plt.figure(figsize=(12, 7))
sns.boxplot(data=df_plot, x="Group", y="Gait_Speed", hue="Task_Condition", palette="viridis")
sns.stripplot(data=df_plot, x="Group", y="Gait_Speed", hue="Task_Condition", dodge=True, color="black", size=3, jitter=True)
plt.title("Comparação da Velocidade de Marcha: Caminhada Normal vs. Tarefa Dupla por Grupo")
plt.xlabel("Grupo")
plt.ylabel("Velocidade de Marcha (m/s)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Condição da Tarefa")
#plt.savefig("gait_speed_dual_task_comparison.png")
plt.show()
plt.close()
print("Gráfico 'gait_speed_dual_task_comparison.png' gerado com sucesso.")
```

## Análise de Correlação entre Idade e Velocidade de Marcha por Grupo

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "physionet.org/files/gaitpdb/1.0.0/demographics.html"

dfs = pd.read_html(file_path, header=0)

df = dfs[0]

df.rename(columns={
    "Speed_01 (m/sec)": "Speed_Normal"
}, inplace=True)

df["Speed_Normal"] = pd.to_numeric(df["Speed_Normal"], errors='coerce')
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')

df.dropna(subset=["Speed_Normal", "Age"], inplace=True)

df_filtered = df[df["Group"].isin(['PD', 'CO'])].copy()

print("\n### Correlação entre Idade e Velocidade de Marcha por Grupo:\n")
for group in df_filtered["Group"].unique():
    group_df = df_filtered[df_filtered["Group"] == group]
    correlation = group_df["Age"].corr(group_df["Speed_Normal"])
    print(f"Grupo: {group}, Correlação (Idade vs. Velocidade Normal): {correlation:.2f}")


plt.figure(figsize=(12, 7))
sns.scatterplot(data=df_filtered, x="Age", y="Speed_Normal", hue="Group", palette="viridis", alpha=0.7)
sns.regplot(data=df_filtered[df_filtered["Group"] == "PD"], x="Age", y="Speed_Normal", color="blue", scatter=False, ci=None, line_kws={"linestyle":"--"})
sns.regplot(data=df_filtered[df_filtered["Group"] == "CO"], x="Age", y="Speed_Normal", color="green", scatter=False, ci=None, line_kws={"linestyle":"--"})
plt.title("Correlação entre Idade e Velocidade de Marcha por Grupo")
plt.xlabel("Idade")
plt.ylabel("Velocidade de Marcha (m/s)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Grupo")
#plt.savefig("age_gait_speed_correlation.png")
plt.show()
plt.close()
print("Gráfico 'age_gait_speed_correlation.png' gerado com sucesso.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

data_dir = "physionet.org/files/gaitpdb/1.0.0/"

column_names = [
    'Time',
    'L_VGRF_1', 'L_VGRF_2', 'L_VGRF_3', 'L_VGRF_4', 'L_VGRF_5', 'L_VGRF_6', 'L_VGRF_7', 'L_VGRF_8',
    'R_VGRF_1', 'R_VGRF_2', 'R_VGRF_3', 'R_VGRF_4', 'R_VGRF_5', 'R_VGRF_6', 'R_VGRF_7', 'R_VGRF_8',
    'Total_L_Foot_Force',
    'Total_R_Foot_Force'
]

output_dir_one_minute = "/content/drive/MyDrive/Joao Victor [TCC]/Foot_L_R_vgrf_images/OneMinute"
if not os.path.exists(output_dir_one_minute):
    os.makedirs(output_dir_one_minute)
output_dir_two_minute = "/content/drive/MyDrive/Joao Victor [TCC]/Foot_L_R_vgrf_images/TwoMinute"
if not os.path.exists(output_dir_two_minute):
    os.makedirs(output_dir_two_minute)

def plot_vgrf_overlay_with_time_limit(file_path, tempo_60_120):
    try:
        df = pd.read_csv(file_path, sep='\\s+', header=None, names=column_names)
        file_name = os.path.basename(file_path).replace(".txt", "")

        # Filtrar o DataFrame com base no limite de tempo em segundos
        df_window = df[df['Time'] <= tempo_60_120].copy()

        plt.figure(figsize=(14, 7))
        sns.lineplot(data=df_window, x='Time', y='Total_L_Foot_Force', label='Pé Esquerdo', color='blue', alpha=0.7)
        sns.lineplot(data=df_window, x='Time', y='Total_R_Foot_Force', label='Pé Direito', color='red', alpha=0.7)

        plt.title(f'Foot_L_R - {file_name} ({tempo_60_120}s)')
        plt.xlabel('Tempo (segundos)')
        plt.ylabel('Força (Newtons)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()

        if tempo_60_120 == 60:
            output_dir = output_dir_one_minute
        elif tempo_60_120 == 120:
            output_dir = output_dir_two_minute
        else:
            output_dir = "/content/drive/MyDrive/Joao Victor [TCC]/Foot_L_R_vgrf_images/CustomTime"
            if not os.path.exists(output_dir):
                 os.makedirs(output_dir)


        output_path = os.path.join(output_dir, f'vgrf_overlay_{file_name}_time_{tempo_60_120}s.png')
        plt.savefig(output_path)

        #plt.show()
        plt.close()
        print(f"Gráfico '{output_path}' gerado com sucesso.")

    except FileNotFoundError:
        print(f"Erro: O arquivo {file_path} não foi encontrado. Certifique-se de que ele está no diretório correto.")
    except Exception as e:
        print(f"Ocorreu um erro ao plotar VGRF para {file_path}: {e}")


for root, dirs, files in os.walk(data_dir):
    for file in files:
        #if file.endswith(".txt"):
        if file.startswith("GaPt"):
            full_path = os.path.join(root, file)
            # Chamando a nova função com limite de tempo em segundos
            plot_vgrf_overlay_with_time_limit(full_path, 60) # Exemplo para 60 segundos
            plot_vgrf_overlay_with_time_limit(full_path, 120) # Exemplo para 120 segundos
```
