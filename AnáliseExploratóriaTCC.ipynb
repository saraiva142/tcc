#### Dados Demográficos

```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns

file_path = 'physionet.org/files/gaitpdb/1.0.0/demographics.html'

dfs = pd.read_html(file_path, header=0)

df = dfs[0] #Só tem uma tabela lá mas p garantir pegar sempre a primeira

print("### Primeiras 5 linhas: \n")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\n### Informações dos dados Demográficoss:\n")
df.info()

print("\n### Estatísticas dos dados Demográficos:\n")
print(df.describe(include='all').to_markdown(numalign="left", stralign="left"))

# Colocando a coluna de Age para numérico
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df.dropna(subset=["Age"], inplace=True)

print("\n### Distribuição de Paciente vs. Controle:\n")
print(df["Group"].value_counts().to_markdown(numalign="left", stralign="left"))

print("\n### Distribuição de Sexo por Grupo:\n")
print(pd.crosstab(df["Group"], df["Gender"]).to_markdown(numalign="left", stralign="left"))

print("\n### Estatística de Idade por Grupo:\n")
print(df.groupby("Group")["Age"].describe().to_markdown(numalign="left", stralign="left"))

# Parte dos gráficoss

#Gráficos de idade por grupo
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Age", hue="Group", kde=True, palette="viridis")
plt.title("Distribuição de Idade por Grupo (Parkinson vs. Controle)")
plt.xlabel("Idade")
plt.ylabel("Contagem")
plt.savefig("age_distribution.png")
print("\n")
plt.show()
#plt.close()
print("Gráfico 'age_distribution.png' baixado na pasta desse colab com sucesso.")

#Gráficos da distribuição de sexo por grupo
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Group", hue="Gender", palette="viridis")
plt.title("Distribuição de Sexo por Grupo (Parkinson vs. Controle)")
plt.xlabel("Grupo")
plt.ylabel("Contagem")
plt.savefig("gender_distribution.png")
print("\n")
plt.show()
#plt.close()
print("Gráfico 'gender_distribution.png' baixado na pasta desse colab com sucesso.")
```

### Análise geral de vgrf

```python
import os

data_dir = "physionet.org/files/gaitpdb/1.0.0/"

column_names = [
    'Time',
    'L_VGRF_1', 'L_VGRF_2', 'L_VGRF_3', 'L_VGRF_4', 'L_VGRF_5', 'L_VGRF_6', 'L_VGRF_7', 'L_VGRF_8',
    'R_VGRF_1', 'R_VGRF_2', 'R_VGRF_3', 'R_VGRF_4', 'R_VGRF_5', 'R_VGRF_6', 'R_VGRF_7', 'R_VGRF_8',
    'Total_L_Foot_Force',
    'Total_R_Foot_Force'
]

def analyze_and_plot_vgrf(file_path):

  df = pd.read_csv(file_path, sep='\s+', header=None, names=column_names)

  file_name = os.path.basename(file_path).replace('.txt', '')

  # Plotar a série temporal da força de reação vertical do solo (VGRF) para o pé esquerdo
  plt.figure(figsize=(12, 6))
  sns.lineplot(data=df, x='Time', y='Total_L_Foot_Force', color='blue')
  plt.title(f'Série Temporal da Força de Reação Vertical do Solo (Pé Esquerdo) - {file_name}' )
  plt.xlabel('Tempo (segundos)')
  plt.ylabel('Força (Newtons)')
  plt.grid(True)
  #plt.savefig(f'vgrf_left_foot_timeseries_{file_name}.png') #Salvar nn senão fica mt coisa salva aqui
  plt.show
  #plt.close()
  print(f"Gráfico 'vgrf_left_foot_timeseries_{file_name}.png' gerado com sucesso.")

  # Plotar a série temporal da força de reação vertical do solo (VGRF) para o pé direito
  plt.figure(figsize=(12, 6))
  sns.lineplot(data=df, x='Time', y='Total_R_Foot_Force', color='red')
  plt.title(f'Série Temporal da Força de Reação Vertical do Solo (Pé Direito) - {file_name}' )
  plt.xlabel('Tempo (segundos)')
  plt.ylabel('Força (Newtons)')
  plt.grid(True)
  #plt.savefig(f'vgrf_right_foot_timeseries_{file_name}.png')
  plt.show()
  #plt.close()
  print(f"Gráfico 'vgrf_right_foot_timeseries_{file_name}.png' gerado com sucesso.")

analyze_and_plot_vgrf(os.path.join(data_dir, "GaCo01_01.txt"))
```

```python
#tentar análisar tudo de uma vez já que está como função kk
##Mas lembra que tem mt arquivo kk

for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.startswith("GaCo02_02.txt"):
            file_path = os.path.join(root, file)
            analyze_and_plot_vgrf(file_path)
```

## Tentando fazer as seguinte análise:
### A comparação das velocidades de marcha entre os grupos de Parkinson e Controle

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "physionet.org/files/gaitpdb/1.0.0/demographics.html"

dfs = pd.read_html(file_path, header=0)

df = dfs[0]

df.rename(columns={'Speed_01 (m/sec)': 'Speed_01'}, inplace=True)

#tratar p essa coluna de velocidade ser numérica para fazer as estatísticas
df["Speed_01"] = pd.to_numeric(df["Speed_01"], errors='coerce')
df.dropna(subset=["Speed_01"], inplace=True)

#Pegar só os PD e Controles
df_filtered = df[df['Group'].isin(['PD', 'CO'])]

print("### Estatísticas das velocidades de marcha por grupo:\n")
print(df_filtered.groupby("Group")["Speed_01"].describe().to_markdown(numalign="left", stralign="left"))

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_filtered, x="Group", y="Speed_01", palette="viridis")
sns.stripplot(data=df_filtered, x="Group", y="Speed_01", color="black", size=3, jitter=True)
plt.title("Distribuição da Velocidade de Marcha por Grupo (Parkinson vs. Controle)")
plt.xlabel("Grupo")
plt.ylabel("Velocidade de marcha (m/s)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
#plt.savefig("speed_distribution.png")
print("\n")
plt.show()
#plt.close()
print("Gráfico 'speed_distribution.png' baixado na pasta desse colab com sucesso.")
```
